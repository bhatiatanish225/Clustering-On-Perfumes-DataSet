```python
from pycaret.utils import version
version()
```

```python
from pycaret.datasets import get_data
dataset =get_data('index')
```

```python
perfume_dataset=get_data('perfume')
```

```python
from pycaret import clustering
from pycaret.clustering import *
clusteringparameters=setup(perfume_dataset)
```

```python
clusteringmodel = create_model('kmeans', num_clusters=4)
```

```python
predictions=assign_model(clusteringmodel)
predictions
```

```python
x=save_model(clusteringmodel,'clusteringmodelfile')
```

```python
from google.colab import files
files.download('clusteringmodelfile.pkl')
```

```python
newdata=get_data('perfume').iloc[:10]
```

```python
newpredictions=predict_model(clusteringmodel,data=newdata)
newpredictions
```
